```python
import time
import random

def lower_bound(arr, x):
    left, right = 0, len(arr) - 1
    lower = 0  

    while left <= right:
        mid = (left + right) // 2
        if arr[mid] < x:
            left = mid + 1
        elif arr[mid] > x:
            right = mid - 1
        else:
            lower = mid
            right = mid - 1  

    return lower

def upper_bound(arr, x):
    left, right = 0, len(arr) - 1
    upper = 0  

    while left <= right:
        mid = (left + right) // 2
        if arr[mid] < x:
            left = mid + 1
        elif arr[mid] > x:
            right = mid - 1
        else:
            upper = mid
            left = mid + 1 

    return upper


def measure_time():
    sizes = [100, 1000, 10000, 50000]
    results = []

    for size in sizes:
        arr = sorted(random.randint(0, size // 2) for _ in range(size))
        x = random.choice(arr)  
        total_time = 0
        iterations = 100

        for _ in range(iterations):
            start_time = time.time()
            lower_bound(arr, x)
            upper_bound(arr, x)
            total_time += time.time() - start_time

        avg_time = total_time / iterations
        results.append((size, avg_time))

    return results


execution_results = measure_time()
print("Среднее время выполнения поиска индексов:", execution_results)
```

```python
def closest_sum(x, y, q):
    i, j = 0, len(y) - 1  
    min_diff = float("inf")
    closest_sum_value = 0
    best_i, best_j = -1, -1

    while i < len(x) and j >= 0:
        s = x[i] + y[j]
        diff = abs(s - q)

        if diff < min_diff:
            min_diff = diff
            closest_sum_value = s
            best_i, best_j = i, j

        if s < q:
            i += 1  
        else:
            j -= 1  

    return x[best_i], y[best_j], closest_sum_value


def measure_time():
    sizes = [100, 1000, 10000, 50000]
    results = []

    for size in sizes:
        x = sorted(random.randint(0, size) for _ in range(size))
        y = sorted(random.randint(0, size) for _ in range(size))
        q = random.randint(0, 2 * size)
        total_time = 0
        iterations = 100

        for _ in range(iterations):
            start_time = time.time()
            closest_sum(x, y, q)
            total_time += time.time() - start_time

        avg_time = total_time / iterations
        results.append((size, avg_time))

    return results

# Запуск замеров
execution_results = measure_time()
print("Среднее время выполнения поиска ближайшей суммы:", execution_results)
```
