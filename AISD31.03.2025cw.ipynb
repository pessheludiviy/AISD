```python
import time #31/03/2025
import random
import matplotlib.pyplot as plt

def perf_sqr(n):
    if n < 0:
        return False  
    left, right = 0, n
    while left <= right:
        mid = (left + right) // 2
        square = mid**2
        if square == n:
            return True
        elif square < n:
            left = mid + 1
        else:
            right = mid - 1
    return False
```

```python
def timemeasure(ranges, iterations=100):
    times = []
    for max_value in ranges:
        total_time = 0
        for _ in range(iterations):
            n = random.randint(0, max_value)
            start_time = time.time()
            perf_sqr(n)
            total_time += time.time() - start_time
        average_time = total_time / iterations
        times.append((max_value, average_time))
    return times


ranges = [10, 100, 1_000, 10_000, 100_000, 1_000_000]
results = timemeasure(ranges)
```

```python
def plot_results(results):
    sizes, times = zip(*results)
    plt.plot(sizes, times, marker="o", label="Бинарный поиск")
    plt.xlabel("Максимальное значение n")
    plt.ylabel("Среднее время выполнения (секунды)")
    plt.title("Зависимость времени выполнения алгоритма от входных данных")
    plt.xscale("log")
    plt.grid(True)
    plt.legend()
    plt.show()

plot_results(results)
```
