```python
from collections import deque
import networkx as nx
import matplotlib.pyplot as plt

def dfs(graph, start):
    visited = set()
    stack = [start]

    while stack:
        node = stack.pop()
        if node not in visited:
            print(node, end=" ")  # Вывод текущей вершины
            visited.add(node)
            stack.extend(reversed(graph[node]))  # Добавляем соседей в стек


graph = {
    'A': ['B', 'C'],
    'B': ['A', 'D', 'E'],
    'C': ['A', 'F', 'G'],
    'D': ['B'],
    'E': ['B', 'H'],
    'F': ['C'],
    'G': ['C'],
    'H': ['E']
}

#Функция общей визуализации
def draw_graph(graph):
    G = nx.Graph()
    for node, neighbors in graph.items():
        for neighbor in neighbors:
            G.add_edge(node, neighbor)

    pos = nx.spring_layout(G)  
    nx.draw(G, pos, with_labels=True, node_color='lightblue', edge_color='gray', node_size=2000, font_size=14)
    plt.title("Визуализация графа")
    plt.show()

#Визуализация графа для DFS метода
def dfs_visual(graph, start):
    visited = set()
    stack = [start]
    
    G = nx.Graph()
    for node, neighbors in graph.items():
        for neighbor in neighbors:
            G.add_edge(node, neighbor)
    
    pos = nx.spring_layout(G)  
    step = 0

    while stack:
        node = stack.pop()
        if node not in visited:
            visited.add(node)
            stack.extend(reversed(graph[node]))

            plt.figure(figsize=(6, 4))
            nx.draw(G, pos, with_labels=True, node_color=['lightgreen' if n in visited else 'lightblue' for n in G.nodes()], edge_color='gray', node_size=2000, font_size=14)
            plt.title(f"DFS: Шаг {step}, посещена вершина {node}")
            plt.show()
            step += 1




draw_graph(graph)
print("Визуализация DFS:")
dfs_visual(graph, 'A')


print("DFS обход:")
dfs(graph, 'A')
```

```python
def bfs(graph, start):
    visited = set()
    queue = deque([start])

    while queue:
        node = queue.popleft()
        if node not in visited:
            print(node, end=" ")  
            visited.add(node)
            queue.extend(graph[node])  

#Накидываю визуализацию графов
def bfs_visual(graph, start):
    visited = set()
    queue = deque([start])

    G = nx.Graph()
    for node, neighbors in graph.items():
        for neighbor in neighbors:
            G.add_edge(node, neighbor)

    pos = nx.spring_layout(G)
    step = 0

    while queue:
        node = queue.popleft()
        if node not in visited:
            visited.add(node)
            queue.extend(graph[node])

            plt.figure(figsize=(6, 4))
            nx.draw(G, pos, with_labels=True, node_color=['lightcoral' if n in visited else 'lightblue' for n in G.nodes()], edge_color='gray', node_size=2000, font_size=14)
            plt.title(f"BFS: Шаг {step}, посещена вершина {node}")
            plt.show()
            step += 1

print("Визуализация BFS:")
bfs_visual(graph, 'A')


print("\nBFS обход:")
bfs(graph, 'A')
```
