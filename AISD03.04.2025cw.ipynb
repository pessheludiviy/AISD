```python
import time #03/04/2025
import random
import matplotlib.pyplot as plt

def count_pp(n, p):
    count = 0
    while n > 0:
        count += n // p
        n //= p
    return count

def product_module(n, p):
    result = 1
    for i in range(1, n + 1):
        if i % p != 0: 
            result = (result * i) % p
    return result

def factorial_module(N, P):
    if N >= P:
        return 0

    power = count_pp(N, P)  
    fact_mod = product_module(N, P)  

    
    return (fact_mod * pow(P, power, P)) % P

def avg_timemeasure(ranges, P, iterations=100):
    results = []
    for N in ranges:
        total_time = 0
        for _ in range(iterations):
            start_time = time.time()
            factorial_module(N, P)
            total_time += time.time() - start_time
        average_time = total_time / iterations
        results.append((N, average_time))
    return results
```

```python
ranges = [10, 100, 1000, 10000, 50000, 100000]

P = int(input("Введите целое неотрицательное число:"))

results = avg_timemeasure(ranges, P)
```

```python
def plot_results(results):
    sizes, times = zip(*results)
    plt.plot(sizes, times, marker="o", label=f"N! mod {P}")
    
    plt.xlabel("Значение N")
    plt.ylabel("Среднее время выполнения (секунды)")
    plt.title("Зависимость времени выполнения от N")
    plt.xscale("log")
    plt.grid(True)
    plt.legend()
    plt.show()

# Построение графика
plot_results(results)
```
