```python
import numpy as np
import time
import matplotlib.pyplot as plt
```

```python
def has_duplicates_set(lst):
    return len(lst) != len(set(lst))

def measure_duplicates_with_set():
    results = []
    for size in [10, 100, 1000, 5000, 10000]:
        lst = [i % 100 for i in range(size)]  # Генерируем список с дубликатами
        start_time = time.time()
        has_duplicates_set(lst)
        end_time = time.time()
        results.append((size, end_time - start_time))
    return results

duplicates_set_results = measure_duplicates_with_set()



def has_duplicates_naive(lst):
    for i in range(len(lst)):
        for j in range(i + 1, len(lst)):
            if lst[i] == lst[j]:
                return True
    return False

def measure_duplicates_naive():
    results = []
    for size in [10, 100, 1000, 5000, 10000]:
        lst = [i % 100 for i in range(size)]  # Генерируем список с дубликатами
        start_time = time.time()
        has_duplicates_naive(lst)
        end_time = time.time()
        results.append((size, end_time - start_time))
    return results

duplicates_naive_results = measure_duplicates_naive()


def intersection_with_set(lst1, lst2):
    return list(set(lst1) & set(lst2))

def measure_intersection_with_set():
    results = []
    for size in [10, 100, 1000, 5000, 10000]:
        lst1 = [i for i in range(size)]
        lst2 = [i for i in range(size // 2, size + size // 2)]
        start_time = time.time()
        intersection_with_set(lst1, lst2)
        end_time = time.time()
        results.append((size, end_time - start_time))
    return results

intersection_set_results = measure_intersection_with_set()

def intersection_naive(lst1, lst2):
    result = []
    for i in lst1:
        if i in lst2:
            result.append(i)
    return result

def measure_intersection_naive():
    results = []
    for size in [10, 100, 1000, 5000, 10000]:
        lst1 = [i for i in range(size)]
        lst2 = [i for i in range(size // 2, size + size // 2)]
        start_time = time.time()
        intersection_naive(lst1, lst2)
        end_time = time.time()
        results.append((size, end_time - start_time))
    return results

intersection_naive_results = measure_intersection_naive()

def remove_duplicates_set(lst):
    return list(set(lst))

def measure_remove_duplicates_with_set():
    results = []
    for size in [10, 100, 1000, 5000, 10000]:
        lst = [i % 100 for i in range(size)]
        start_time = time.time()
        remove_duplicates_set(lst)
        end_time = time.time()
        results.append((size, end_time - start_time))
    return results

remove_duplicates_set_results = measure_remove_duplicates_with_set()

def remove_duplicates_naive(lst):
    result = []
    for i in lst:
        if i not in result:
            result.append(i)
    return result

def measure_remove_duplicates_naive():
    results = []
    for size in [10, 100, 1000, 5000, 10000]:
        lst = [i % 100 for i in range(size)]
        start_time = time.time()
        remove_duplicates_naive(lst)
        end_time = time.time()
        results.append((size, end_time - start_time))
    return results

remove_duplicates_naive_results = measure_remove_duplicates_naive()

def first_duplicate_set(lst):
    seen = set()
    for i in lst:
        if i in seen:
            return i
        seen.add(i)
    return None

def measure_first_duplicate_with_set():
    results = []
    for size in [10, 100, 1000, 5000, 10000]:
        lst = [i % 100 for i in range(size)]
        start_time = time.time()
        first_duplicate_set(lst)
        end_time = time.time()
        results.append((size, end_time - start_time))
    return results

first_duplicate_set_results = measure_first_duplicate_with_set()

def first_duplicate_naive(lst):
    for i in range(len(lst)):
        for j in range(i + 1, len(lst)):
            if lst[i] == lst[j]:
                return lst[i]
    return None

def measure_first_duplicate_naive():
    results = []
    for size in [10, 100, 1000, 5000, 10000]:
        lst = [i % 100 for i in range(size)]
        start_time = time.time()
        first_duplicate_naive(lst)
        end_time = time.time()
        results.append((size, end_time - start_time))
    return results

first_duplicate_naive_results = measure_first_duplicate_naive()

def two_sum_with_set(lst, target):
    seen = set()
    for num in lst:
        complement = target - num
        if complement in seen:
            return (num, complement)
        seen.add(num)
    return None

def measure_two_sum_with_set():
    results = []
    for size in [10, 100, 1000, 5000, 10000]:
        lst = [i for i in range(size)]
        target = size // 2 + size // 3
        start_time = time.time()
        two_sum_with_set(lst, target)
        end_time = time.time()
        results.append((size, end_time - start_time))
    return results

two_sum_set_results = measure_two_sum_with_set()

def two_sum_naive(lst, target):
    for i in range(len(lst)):
        for j in range(i + 1, len(lst)):
            if lst[i] + lst[j] == target:
                return (lst[i], lst[j])
    return None

def measure_two_sum_naive():
    results = []
    for size in [10, 100, 1000, 5000, 10000]:
        lst = [i for i in range(size)]
        target = size // 2 + size // 3
        start_time = time.time()
        two_sum_naive(lst, target)
        end_time = time.time()
        results.append((size, end_time - start_time))
    return results

two_sum_naive_results = measure_two_sum_naive()


def measure_time(func, lst, repeats=5):
    times = []
    result = None
    for _ in range(repeats):
        start_time = time.time()
        result = func(lst)
        end_time = time.time()
        times.append(end_time - start_time)
    return result, np.mean(times)


sizes = [10, 100, 500, 1000, 2000, 5000, 10_000]
times_set = []
times_naive = []
```
