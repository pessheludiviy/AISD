```python
import time
import random
import matplotlib.pyplot as plt

def measure_time(func, iterations=100, max_value=1_000_000):
    times = []
    for _ in range(iterations):
        a = random.randint(1, max_value)
        b = random.randint(1, max_value)
        start_time = time.time()
        func(a, b)
        times.append(time.time() - start_time)
    return sum(times) / len(times)
```

```python
gcd_subtraction_result = []

def gcd_subtraction(a, b):
    while a != 0 and b != 0:
        if a > b:
            a -= b
        else:
            b -= a
    gcd_subtraction_result.append(a or b)
    return a or b
    
gcd_mod_result = []

def gcd_mod(a, b):
    while b != 0:
        a, b = b, a % b
    gcd_mod_result.append(a)
    return a

def prime_factors(n):
    factors = []
    divisor = 2
    while n > 1:
        while n % divisor == 0:
            factors.append(divisor)
            n //= divisor
        divisor += 1
    return factors

gcd_prime_factors_result = []

def gcd_prime_factors(a, b):
    factors_a = prime_factors(a)
    factors_b = prime_factors(b)
    common_factors = set(factors_a) & set(factors_b)
    gcd = 1
    for factor in common_factors:
        gcd *= factor ** min(factors_a.count(factor), factors_b.count(factor))
    gcd_prime_factors_result.append(gcd)
    return gcd
    
```

```python
ranges = [10, 100, 1000, 10000, 100000, 1000000]

results_subtraction = []
results_mod = []
results_prime_factors = []

for r in ranges:
    print(f"Замер времени выполнения для чисел до {r}")
    results_subtraction.append(measure_time(gcd_subtraction, max_value=r))
    results_mod.append(measure_time(gcd_mod, max_value=r))
    results_prime_factors.append(measure_time(gcd_prime_factors, max_value=r))
```

```python
def plot_results(ranges, results_subtraction, results_mod, results_prime_factors):
    plt.plot(ranges, results_subtraction, label="НОД Вычитание", marker="s")
    plt.plot(ranges, results_mod, label="НОД Mod",  marker="*", linestyle = ":")
    plt.plot(ranges, results_prime_factors, label="НОД Разложение на множетели", marker="o", linestyle = "--")

    plt.xlabel("Максимальное значение чисел")
    plt.ylabel("Среднее время выполнения (секунды)")
    plt.title("Сравнение времени выполнения алгоритмов НОД")
    plt.legend()
    plt.grid(True, linestyle = "--")
    plt.xscale("log")  
    plt.show()


plot_results(ranges, results_subtraction, results_mod, results_prime_factors)
```
