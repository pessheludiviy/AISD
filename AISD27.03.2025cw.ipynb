```python
#27/03/2025
import time
import random
import matplotlib.pyplot as plt
```

```python
def binary_gcd(a, b):
    if a == 0:
        return b
    if b == 0:
        return a

    shift = 0 
    while (a & 1) == 0 and (b & 1) == 0:
        a >>= 1  
        b >>= 1  
        shift += 1  


    while a != b:
        if a < b:
            a, b = b, a 

        if (a & 1) == 0:
            a >>= 1  
        elif (b & 1) == 0:
            b >>= 1 
        else:
            a = (a - b) >> 1  

    return a << shift  
    
#пример работы функции
a = 48
b = 18
print(f"НОД({a}, {b}) = {binary_gcd(a, b)}")
```

```python
def measure_gcd_time():
    ranges = [10, 100, 1000, 10000, 100000, 1000000]
    results = []
    
    for r in ranges:
        total_time = 0
        iterations = 100  
        
        for i in range(iterations):
            a = random.randint(1, r)
            b = random.randint(1, r)
            print(f"Итерация: {i}, число №1: {a}, число №2: {b}, выборка от 1 до {r}")
            start_time = time.time()
            binary_gcd(a, b)
            total_time += time.time() - start_time
            
        results.append((r, total_time / iterations))
    return results


gcd_results = measure_gcd_time()
```

```python
def plot_results(results):
    input_sizes, times = zip(*results)
    plt.plot(input_sizes, times, marker="o", label="Binary GCD")
    plt.xlabel("Максимальное значение входных данных")
    plt.ylabel("Среднее время выполнения (секунды)")
    plt.title("Время выполнения алгоритма Binary GCD")
    plt.xscale("log")
    plt.grid(True)
    plt.legend()
    plt.show()

plot_results(gcd_results)
```
